```python
data = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data),index_col=0)
```

```python
pd.read_csv(StringIO(data),index_col=3) ## setting different column as index
```

```python
data = ('a,b,c\n'
           '4,apple,bat,\n'
            '8,orange,cow,')
# now index not given
```

```python
pd.read_csv(StringIO(data))
```

```python
# But 4 and 8 is my data not index
# to solve
```

```python
pd.read_csv(StringIO(data),index_col=False)
# index will be given as default
```

```python
data = ('a,b,c\n'
           '4,apple,bat,\n'
            '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data), usecols=['b', 'c'],index_col=False)
## Only specified columns will be used
```

```python
## Quoting and Escape Characters¶. Very useful in NLP

data = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python
pd.read_csv(StringIO(data),escapechar='\\')
```

```python
pd.read_csv(StringIO(data))
```
