```python
df.corr()  # here only size tip and total bil is showing.
# the reason is corelation only take floar or int type data.
```

```python
#Most corelated on point 1
```

#  JOIN  PLOT

```python
# A join plot allows to study the relationship between 2
# numeric variables. The central chart display their 
# correlation. It is usually a scatterplot, a hexbin plot, 
# a 2D histogram or a 2D density plot
```

## Univariate Analysis

```python
sns.jointplot (x='tip',y='total_bill',data=df,kind='reg')
```

```python
# basically histogram also shown there
```

```python
sns.jointplot (x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot (x='tip',y='total_bill',data=df,kind='resid')
```

```python
# so most tip was given in position 2
```

```python
# More than Two features we use Pairplot
```

# Pair Plot

```python
#A “pairs plot” is also known as a scatterplot, in which 
#one variable in the same data row is matched with another 
#variable's value, like this: Pairs plots are just 
#elaborations on this, showing all variables paired with 
#all the other variables
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex') # telling one extra feature
```

```python
sns.pairplot(df,hue='smoker') # telling one extra feature
```

```python
df['smoker'].value_counts()
```

# Categorical Plots

```python
# Categorical Plots
# Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

# boxplot
# violinplot
# countplot
# bar plot
```

```python
# They mainly work on categorical values
```

```python
df.dtypes
```

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(x ='sex', data = df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.countplot(x='day',data=df)
```

# Bar Plot

```python
## Bar plot @ in categorical plot x,or y any of them required ,
# as second parameter will be count but in bar plot both x and y are
# required
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

# Boxplot ( Also in matplotlib)

```python
# A box and whisker plot (sometimes called a boxplot) is a 
# graph that presents information from a five-number summary.
```

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day', y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v') # dots in graph are outlines
# this iss for all types
```

```python
sns.boxplot(x="total_bill", y="day", hue="smoker",data=df,palette='rainbow')
```

```python
# What is happening here is that using smoker to show the percentile
# of total_bill and day
```

```python
sns.boxplot(x="total_bill", y="day", hue="smoker",data=df,color='g')
```

#  violin plot

```python
#Violin plot helps us to see both the distribution of data in 
#terms of Kernel density estimation and the box plot
```

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```

```python
sns.violinplot(x='total_bill',y='smoker',data=df,palette='rainbow')
```

```python
# x or y must be numeric type 
```
