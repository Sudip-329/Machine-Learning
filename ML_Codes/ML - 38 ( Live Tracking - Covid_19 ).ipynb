```python
import pandas as pd
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df
```

```python
df = df.iloc[:-1,1:]
```

```python
df.tail()
```

```python
df # as 36 no row and number 1 col(sno) not needed
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states = np.arange(len(df['state_name'])) 
```

```python
total_states
```

```python
df['active'] # as data taken in 2023 so ative cases are very low
```

```python
max(df['positive'])
```

 # Total Positive Cases Based On States In India

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['positive'], align='center', alpha=0.5)
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(8,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['positive'], align='center', alpha=0.5, color=(1,0,0), edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states, df['state_name'])
plt.xlim(1,max(df['positive'])+100) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show() 
```

## Active Number of Cases based on States

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['active'], align='center', alpha=0.5,  
                 color=(1,0.5,0),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['active'])+10) 
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show() 
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['new_active'], align='center', alpha=0.5,  
                 color=(1,0.5,0),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['new_active'])+10) 
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show() 
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(10,8), dpi=80, facecolor='r', edgecolor='b')
plt.barh(total_states,df['death'],align='center', alpha=0.5,color=(0,0,1),edgecolor=(0.5,0.4,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['death'])+10)
plt.xlabel("Death Number of Cases")
plt.ylabel("Corona Virus Cases")
plt.show()
```
